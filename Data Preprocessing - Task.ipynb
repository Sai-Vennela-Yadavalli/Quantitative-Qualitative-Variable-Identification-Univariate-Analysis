```python
# import neccessary libraries

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset

Dataset = pd.read_csv("kidney_disease.csv")

Dataset
```

```python
# Checking is there any null / missing values

Dataset.isna().sum()
```

```python
from Univariate_codings import univariate
```

```python
Dataset[Quan].isna().sum()
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan, strategy = "mean")
imp.fit_transform(Dataset[Quan])
data_table = imp.transform(Dataset[Quan])
data_table = pd.DataFrame(data_table, columns = Quan)
```

```python
data_table[Quan].isna().sum()
```
